```python
import pandas as pd 
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data
RD = pd.read_csv('RDNorm.csv', names =['Index','D Time','D Interval'])
RA = pd.read_csv('RANorm.csv', names =['Index','A Time','A Interval'])
LD = pd.read_csv('LDNorm.csv', names =['Index','D Time','D Interval'])
LA = pd.read_csv('LANorm.csv', names =['Index','A Time','A Interval'])

#Convert interval string to seconds
RD["D Interval"] = pd.to_timedelta(RD["D Interval"].radd("00:")).dt.total_seconds()
RA["A Interval"] = pd.to_timedelta(RA["A Interval"].radd("00:")).dt.total_seconds()
LD["D Interval"] = pd.to_timedelta(LD["D Interval"].radd("00:")).dt.total_seconds()
LA["A Interval"] = pd.to_timedelta(LA["A Interval"].radd("00:")).dt.total_seconds()
```

```python
LD = pd.read_csv('LDNorm.csv', names =['Index','D Time','D Interval'])
LD["D Intx`xerval"] = pd.to_timedelta(LD["D Interval"].radd("00:")).dt.total_seconds()


RD = pd.read_csv('RDNorm.csv', names =['Index','D Time','D Interval'])
RD["D Interval"] = pd.to_timedelta(RD["D Interval"].radd("00:")).dt.total_seconds()
```

```python
data = LA["A Interval"]
```

```python

sns.set_style('whitegrid')
fig, ax1 = plt.subplots()
sns.histplot(data, kde=False, ax=ax1)
ax2 = ax1.twinx()
sns.kdeplot(data ,ax=ax2, clip=(0, None))



ax1.set_xlabel('Value')
ax1.set_ylabel('Count')
ax2.set_ylabel('Density')
ax2.set_ylim(bottom=0)  


plt.show()
```
